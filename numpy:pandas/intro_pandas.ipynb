# 13.Use map and the following dictionary to replace the genres.

```python
dic = {1:'Sports',2:'Platform',3:'Racing',4:'Role-Playing',5:'Puzzle',6:'Misc',
       7:'Shooter',8:'Simulation',9:'Action',10:'Fighting',11:'Adventure',12:'Strategy'}

def replacement_dic(x):
    for key in dic.keys():
        if key==x:
            return dic[key]

df2['Genre']=list(map(replacement_dic,df2['Genre']))
df2
```

# 14.Which action game has the highest global sales?

```python
df2[df2['Genre']=='Action'].groupby('Name').sum().sort_values('Global_Sales',ascending=False)['Global_Sales'].index[0]
```
