```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x = 'day', y= 'total_bill', hue='time', palette=['m','g'], data=df)
None
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='day', y = 'tip', hue='smoker', data=df)
None
```

```python
sns.regplot(data = df, x = 'total_bill', y = 'tip')
```

```python
sns.scatterplot(data = df, x = 'total_bill', y = 'tip', hue='size', size='size', hue_norm= (0,7), sizes=(25, 300))
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(data = df, x = 'total_bill', y = 'tip', hue='size', size='size', hue_norm= (0,7), sizes=(25, 300))
sns.regplot(data = df, x = 'total_bill', y = 'tip')
```

```python
plt.figure(figsize=(15,5))
sns.pairplot(df, hue= 'smoker', palette = "rocket")
```

```python
g = sns.FacetGrid(df, col = "time")
g.map(plt.hist, "tip")
plt.show()
```

```python
g = sns.FacetGrid(df, col = "sex", hue = "smoker")
g.map(plt.scatter, "total_bill", "tip")
plt.show()
```
